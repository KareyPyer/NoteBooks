```python
from collections import Counter
import numpy as np
import matplotlib.pyplot as plt

def look_and_say(s):
    """Génère le terme suivant"""
    result = []
    i = 0
    while i < len(s):
        count = 1
        while i + 1 < len(s) and s[i] == s[i+1]:
            i += 1
            count += 1
        result.append(f"{count}{s[i]}")
        i += 1
    return "".join(result)

# Génération des termes
terms = ["1"]
for _ in range(20):
    terms.append(look_and_say(terms[-1]))

lengths = [len(t) for t in terms]
print(f"Terme 20 : {terms[20][:50]}...")
print(f"Longueurs : {lengths}")

# Estimation de la constante de Conway
ratios = [lengths[i]/lengths[i-1] for i in range(2, len(lengths))]
conway_constant = np.mean(ratios[-5:])
print(f"\n🎯 Constante de Conway estimée : {conway_constant:.6f}")
print(f"   (Valeur réelle ≈ 1.303577)")

# Visualisation
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.semilogy(range(len(lengths)), lengths, 'bo-', markersize=8)
ax1.set_xlabel("n")
ax1.set_ylabel("Longueur du terme n")
ax1.set_title("📈 Croissance exponentielle de la suite de Conway")
ax1.grid(True, alpha=0.3)

ax2.plot(range(2, len(lengths)), ratios, 'r.-', markersize=10)
ax2.axhline(y=1.303577, color='g', linestyle='--', label='λ ≈ 1.303577')
ax2.set_xlabel("n")
ax2.set_ylabel("Ratio longueur(n)/longueur(n-1)")
ax2.set_title("🎯 Convergence vers la constante de Conway")
ax2.legend()
ax2.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Analyse des fréquences des "éléments atomiques" de Conway
# Conway a démontré que la suite se décompose en 92 "éléments" (comme le tableau périodique !)
print("\n🔬 Décomposition en éléments de Conway (92 éléments 'atomiques')")
print("   Exemple : '22' = 'deux 2' est l'élément 'H' (Hydrogène de Conway)")
```
