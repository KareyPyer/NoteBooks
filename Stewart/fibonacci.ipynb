```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle

def fibonacci_spiral(n_points=500, golden_angle=137.508):
    """Modèle phyllotaxique de Stewart — angle d'or = 137.508°"""
    golden_rad = np.radians(golden_angle)
    r = np.sqrt(np.arange(n_points))
    theta = np.arange(n_points) * golden_rad
    
    fig, ax = plt.subplots(figsize=(10, 10), subplot_kw=dict(projection='polar'))
    colors = plt.cm.viridis(np.linspace(0, 1, n_points))
    
    ax.scatter(theta, r, c=colors, s=50, alpha=0.8)
    ax.set_title("🌻 Phyllotaxie — Spirales de Fibonacci dans la Nature\n"
                 f"Angle d'or = {golden_angle}°", fontsize=14, pad=20)
    ax.set_rticks([])
    ax.set_xticks([])
    plt.tight_layout()
    plt.show()
    
    # Analyse des spirales paraboliques
    return r, theta

r, theta = fibonacci_spiral(800)

# Vérification : les spirales correspondent aux nombres de Fibonacci consécutifs
fib = [1, 1]
for _ in range(15):
    fib.append(fib[-1] + fib[-2])

print("🔢 Nombres de Fibonacci :", fib[:12])
print("🌿 Les spirales visibles sur un tournesol sont généralement des paires")
print("   de nombres de Fibonacci consécutifs (ex: 21 et 34, ou 34 et 55)")
```
