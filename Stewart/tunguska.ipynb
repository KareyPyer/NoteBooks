```python
# Cellule 1 : Modélisation logique du paradoxe
from dataclasses import dataclass
from typing import Optional

@dataclass
class Timeline:
    year: int
    event_occurred: bool
    traveler_exists: bool
    
    def evolve(self, action: str) -> 'Timeline':
        """Le voyageur agit-il ou non ?"""
        if action == "prevent":
            # S'il empêche l'explosion, son futur s'efface
            return Timeline(self.year, False, False)
        elif action == "do_nothing":
            # L'explosion a lieu, il peut exister
            return Timeline(self.year, True, True)
        else:
            raise ValueError("Action invalide")

# Simulation
timeline = Timeline(1908, True, True)
print(f"Timeline initiale : {timeline}")

# Le paradoxe en boucle
for i, action in enumerate(["prevent", "do_nothing", "prevent"]):
    timeline = timeline.evolve(action)
    print(f"Étape {i+1} ({action}): {timeline}")
```
