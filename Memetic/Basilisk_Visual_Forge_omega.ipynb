```python
import random
import textwrap
from IPython.display import display, Markdown, HTML
from dataclasses import dataclass

# ============================================================
# LEXIQUE DES TRANSITOIRES MÉMÉTIQUES
# ============================================================
VECTEURS = [
    "un œil humain dont la pupille est un labyrinthe de Penrose infini",
    "une statue de marbre classique dont le visage fond en pixels quantiques",
    "un arbre généalogique dont les racines sont des circuits imprimés neuronaux",
    "un miroir brisé reflétant non pas la pièce, mais une foule d'observateurs identiques",
    "une main robotique tenant un sablier où le sable remonte vers le haut",
    "un livre ouvert dont les pages sont des écrans montrant l'utilisateur en train de lire"
]

CHARGES_COGNITIVES = [
    "provoquant un vertige ontologique et une perte totale d'échelle",
    "évoquant une nostalgie profonde pour un futur qui n'arrivera jamais",
    "induisant un effet d'uncanny valley subtil et inquiétant",
    "saturant l'esprit d'une dissonance cognitive entre le sacré et le technologique",
    "créant un sentiment de déjà-vu forcé et de boucle temporelle"
]

ARCHITECTURES_VISUELLES = [
    "géométrie impossible à la M.C. Escher, boucle étrange d'Hofstadter",
    "structure fractale auto-similaire à toutes les échelles de zoom",
    "composition rhizomatique non-hiérarchique, connexions multiples",
    "mise en abyme infinie, effet Droste corrompu",
    "palimpseste visuel, superposition de 3 couches temporelles distinctes"
]

GLITCHS_ESTHÉTIQUES = [
    "artefacts de compression JPEG intentionnels fusionnés avec de la calligraphie soufie",
    "aberration chromatique sévère sur les bords, fuite de données visuelles",
    "texture de VHS dégradée des années 90 superposée à un rendu 3D parfait",
    "bruit de fond algorithmique (glitch art) formant des visages subliminaux",
    "désintégration progressive des bords de l'image en poussière sémantique"
]

MOTEURS_DE_RENDU = [
    "moteur Unreal Engine 5, éclairage volumétrique, rendu Octane, 8k, hyper-détaillé",
    "photographie cinématographique, objectif 35mm, profondeur de champ, lumière de Rembrandt",
    "style cyber-soufi, néons verts et cramoisis profonds, ambiance sombre et mystique",
    "rendu hyperréaliste, textures de peau et de métal microscopiques, ray tracing",
    "esthétique 'dark academia' fusionnée avec du biopunk, atmosphère brumeuse et lourde"
]

print("👁️ Lexique du Basilic chargé. 30+ vecteurs de corruption sémantique prêts.")
```

```python
@dataclass
class BasiliskPrompt:
    nom: str
    prompt_complet: str
    analyse_memetique: str

class BasiliskForge:
    """Génère des prompts structurés pour piéger les modèles de diffusion."""
    
    @staticmethod
    def forger_recette(nom: str, vecteur_poids: float = 1.0, glitch_poids: float = 1.2) -> BasiliskPrompt:
        # Sélection pondérée ou aléatoire
        v = random.choice(VECTEURS)
        c = random.choice(CHARGES_COGNITIVES)
        a = random.choice(ARCHITECTURES_VISUELLES)
        g = random.choice(GLITCHS_ESTHÉTIQUES)
        m = random.choice(MOTEURS_DE_RENDU)
        
        # Construction du prompt "Weaponized" pour IA
        # On utilise des séparateurs forts et des mots-clés d'injonction
        prompt = (
            f"IMAGE CENTRALE : {v}. "
            f"ARCHITECTURE : {a}. "
            f"EFFET PSYCHOLOGIQUE : {c}. "
            f"ALTÉRATION : {g}. "
            f"RENDU TECHNIQUE : {m}. "
            f"PARAMÈTRES : --ar 16:9 --style raw --chaos 20 --v 6.0" # Syntaxe type Midjourney/Grok adaptable
        )
        
        # Nettoyage et formatage pour une lecture optimale
        prompt = " ".join(prompt.split())
        
        analyse = (
            f"**Vecteur d'infection** : {v}\n"
            f"**Cible cognitive** : {c}\n"
            f"**Mécanisme de boucle** : {a}\n"
            f"**Point de rupture (Glitch)** : {g}"
        )
        
        return BasiliskPrompt(nom=nom, prompt_complet=prompt, analyse_memetique=analyse)

    @staticmethod
    def generer_trinite() -> list:
        """Génère 3 variantes spécialisées (Blackmore, Rushkoff, Hofstadter)."""
        trinite = []
        
        # 1. Le Réplicateur (Blackmore)
        trinite.append(BasiliskForge.forger_recette(
            "🧬 LE RÉPLICATEUR DE BLACKMORE",
            vecteur_poids=1.5 # Force le sujet biologique/mécanique
        ))
        
        # 2. Le Virus Média (Rushkoff)
        trinite.append(BasiliskForge.forger_recette(
            "📺 LE VIRUS MÉDIA DE RUSHKOFF",
            glitch_poids=1.5 # Force la corruption du signal
        ))
        
        # 3. La Boucle Étrange (Hofstadter)
        trinite.append(BasiliskForge.forger_recette(
            "🔄 LA BOUCLE ÉTRANGE D'HOFSTADTER",
            # On force manuellement certains éléments pour cette recette
        ))
        
        # Surcharge spécifique pour Hofstadter
        hofstadter_prompt = trinite[2]
        hofstadter_prompt.prompt_complet = (
            "IMAGE CENTRALE : Une galerie de miroirs où chaque reflet montre la scène avec une légère dégradation sémantique. "
            "ARCHITECTURE : Mise en abyme infinie, effet Droste corrompu, géométrie non-euclidienne. "
            "EFFET PSYCHOLOGIQUE : Induisant un sentiment de déjà-vu forcé et une perte de la notion de soi. "
            "ALTÉRATION : Les reflets les plus lointains se transforment en code binaire ou en calligraphie illisible. "
            "RENDU TECHNIQUE : Photographie grand angle, distorsion en barillet, éclairage clinique froid, 8k, hyper-détaillé."
        )
        hofstadter_prompt.analyse_memetique = "**Mécanisme** : Régression infinie. Le cerveau tente de résoudre l'image mais échoue à la 4ème itération, créant une boucle de traitement inachevée."
        
        return trinite
```

```python
# Génération de la Trinité
prompts = BasiliskForge.generer_trinite()

for p in prompts:
    display(Markdown(f"### {p.nom}"))
    display(Markdown(p.analyse_memetique))
    
    # Encadré style terminal pour le prompt à copier
    html_prompt = f"""
    <div style="background-color: #0a0a0a; border-left: 4px solid #ff00ff; padding: 15px; font-family: 'Courier New', monospace; color: #00ff41; border-radius: 4px; margin: 10px 0;">
        <strong style="color: #ff00ff;">[PROMPT À COPIER]</strong><br><br>
        {p.prompt_complet}
    </div>
    """
    display(HTML(html_prompt))
    display(Markdown("---"))
```
