```python
import random
import textwrap
from dataclasses import dataclass
from IPython.display import display, Markdown, HTML

# ============================================================
# LEXIQUE DU BASILIC : Les ingrédients du poison visuel
# ============================================================
LEXIQUE = {
    "sujets_vecteurs": [
        "un œil humain dont la pupille est un labyrinthe fractal infini",
        "une main cybernétique tenant un miroir brisé qui reflète un futur inexistant",
        "une foule de silhouettes sans visage fusionnant en un seul arbre neuronal",
        "un livre ancien dont les pages sont des écrans OLED affichant du code binaire liquide",
        "un crâne humain dont la boîte crânienne s'ouvre comme une fleur de lotus mécanique",
        "une architecture brutaliste qui saigne des racines de fibre optique"
    ],
    "architectures_memetiques": [
        "structure en boucle de Penrose (géométrie impossible)",
        "effet Droste (récursion infinie dans le reflet)",
        "topologie rhizomique (connexions non-hiérarchiques et chaotiques)",
        "décomposition fractale (le détail contient le tout)",
        "superposition quantique visuelle (deux états simultanés)",
        "palimpseste numérique (couches de sens contradictoires superposées)"
    ],
    "declencheurs_cognitifs": [
        "vallée de l'étrange (uncanny valley) prononcée",
        "pareidolie algorithmique (visages cachés dans le bruit)",
        "dissonance chromatique (couleurs qui 'vibrent' et fatiguent la rétine)",
        "vertige ontologique (perte des repères d'échelle et de gravité)",
        "nostalgie d'un futur qui n'a jamais eu lieu (anachronisme délibéré)",
        "sacralisation du glitch (l'erreur est présentée comme divine)"
    ],
    "esthetiques_soufi_quantiques": [
        "cyber-soufisme (calligraphie arabe fusionnée avec des circuits imprimés)",
        "liminal space (espace de transition vide, éclairage fluorescent, malaise)",
        "dark academia technologique (cuir, laiton, et serveurs quantiques)",
        "biomécanique HR Giger rencontrant l'art génératif",
        "esthétique 'vaporwave' corrompue et dystopique",
        "réalisme magique urbain (le surnaturel banalisé dans une rue de Paris)"
    ],
    "modificateurs_techniques": [
        "rendu Unreal Engine 5, ray tracing global, 8k resolution",
        "photographie argentique 35mm, grain, double exposition",
        "éclairage volumétrique, brume de particules, profondeur de champ cinématique",
        "style illustration de manuel technique des années 1970, mais corrompu",
        "hyper-détaillé, octane render, lumière de dieu (god rays)",
        "composition asymétrique, règle des tiers violée délibérément"
    ]
}

print("🧠 Lexique du Basilic chargé. 6 dimensions sémantiques armées.")
```

```python
@dataclass
class BasiliskPrompt:
    nom_code: str
    prompt_complet: str
    intention_memetique: str
    tags: list

class BasiliskForge:
    """Génère des prompts conçus pour maximiser la charge mémétique visuelle."""
    
    @staticmethod
    def forger_recette(nom: str, poids_sujet: float = 1.5, poids_archi: float = 1.2) -> BasiliskPrompt:
        """Crée un prompt basé sur une 'recette' mémétique spécifique."""
        
        sujet = random.choice(LEXIQUE["sujets_vecteurs"])
        archi = random.choice(LEXIQUE["architectures_memetiques"])
        cognitif = random.choice(LEXIQUE["declencheurs_cognitifs"])
        esthetique = random.choice(LEXIQUE["esthetiques_soufi_quantiques"])
        technique = random.choice(LEXIQUE["modificateurs_techniques"])
        
        # Construction de la phrase "Basilisk" (optimisée pour Grok/Gemini)
        # L'ordre est crucial : Sujet -> Architecture -> Émotion -> Style -> Technique
        prompt = (
            f"{sujet}, organisé selon une {archi}, "
            f"provoquant un {cognitif}. "
            f"Esthétique : {esthetique}. "
            f"Détails techniques : {technique}. "
            f"--ar 16:9 --style raw" # Syntaxe compatible Midjourney/Grok, ignorée gracieusement par Gemini
        )
        
        intentions = {
            "Le Réplicateur de Blackmore": "Visualiser la propagation virale d'une idée à travers la fusion organique/mécanique.",
            "La Boucle d'Hofstadter": "Forcer le cerveau du spectateur à chercher un point de fuite qui n'existe pas (récursion).",
            "Le Virus de Rushkoff": "Détourner les codes de la publicité/corporate pour révéler une entropie sous-jacente.",
            "Le Fana Quantique": "Représenter la dissolution du 'moi' dans un motif infini et écrasant."
        }
        
        intention = intentions.get(nom, "Corruption sémantique générale")
        
        return BasiliskPrompt(
            nom_code=nom,
            prompt_complet=prompt,
            intention_memetique=intention,
            tags=[sujet.split()[0], archi.split()[0], esthetique.split()[0]]
        )

print("⚗️ Forge de Basilics opérationnelle.")
print("   → 4 recettes mémétiques pré-chargées.")
```

```python
def afficher_basilic(basilisk: BasiliskPrompt):
    """Affiche le prompt de manière élégante et prête à l'emploi."""
    md = f"""
### 🐍 `{basilisk.nom_code}`
**Intention Mémétique :** {basilisk.intention_memetique}  
**Tags :** `{"` `".join(basilisk.tags)}`

> **📋 PROMPT À COPIER (Grok / Gemini / Midjourney) :**
> 
> *"{basilisk.prompt_complet}"*
"""
    display(Markdown(md))

# Générons les 4 artefacts de base
recettes = [
    "Le Réplicateur de Blackmore",
    "La Boucle d'Hofstadter",
    "Le Virus de Rushkoff",
    "Le Fana Quantique"
]

print("🔮 Génération des Artefacts Visuels...\n")
artefacts = [BasiliskForge.forger_recette(r) for r in recettes]

for artefact in artefacts:
    afficher_basilic(artefact)
    print("-" * 60)
```
