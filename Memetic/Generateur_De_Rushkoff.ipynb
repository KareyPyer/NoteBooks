```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from ipywidgets import interact, FloatSlider
import random

# Configuration du style "Industriel / Cyberpunk"
sns.set_theme(style="darkgrid")
plt.rcParams['figure.facecolor'] = '#121212'
plt.rcParams['axes.facecolor'] = '#1e1e1e'
plt.rcParams['text.color'] = '#e0e0e0'
```

```python
def creer_systeme_immunitaire(n_noeuds=500, connexions=3):
    """Génère le réseau social (Barabási-Albert) avec un seuil d'immunité."""
    G = nx.barabasi_albert_graph(n_noeuds, connexions)
    
    # Le système immunitaire : plus un nœud est connecté (influencer), 
    # plus il est résistant aux mèmes "basiques" (weaponized memes).
    for node in G.nodes():
        degre = G.degree(node)
        # Seuil d'immunité proportionnel à l'influence (le cynisme des masses)
        G.nodes[node]['immunite'] = np.log1p(degre) * np.random.uniform(0.5, 1.5)
        G.nodes[node]['etat'] = 'Sain' # S, I, R (Susc, Infecté, Réfractaire)
        
    return G

G_reseau = creer_systeme_immunitaire()
print(f"Système immunitaire généré : {G_reseau.number_of_nodes()} hôtes.")
```

```python
class RushkoffVirus:
    def __init__(self, resonance_narrative, taux_mutation, bruit_industriel):
        """
        resonance_narrative: La capacité du mème à s'intégrer dans les mythes existants.
        taux_mutation: La capacité du mème à changer de forme pour éviter la censure.
        bruit_industriel (Facteur Reznor): Le glitch, la dissonance cognitive, 
                                           l'émotion brute qui bypass le cortex préfrontal.
        """
        self.resonance = resonance_narrative
        self.mutation = taux_mutation
        self.noise = bruit_industriel
        self.generation = 0
        self.code_source = "MEME_WEAPONIZED_V1"

    def muter(self):
        """Le virus s'adapte au système immunitaire."""
        self.generation += 1
        self.code_source = f"MEME_MUTANT_GEN_{self.generation}"
        # La mutation augmente la pénétration mais peut réduire la résonance
        self.resonance *= np.random.uniform(0.8, 1.2)
        self.noise *= np.random.uniform(0.9, 1.3)
        return self.code_source

    def calculer_force_infection(self):
        # La force du virus n'est pas sa violence, mais sa capacité à se cacher 
        # dans le bruit de fond (le "Drop-out" de Rushkoff).
        return (self.resonance * 0.6) + (self.noise * 0.4)
```

```python
def propager_virus(reseau, virus_zero, t_max=20):
    """Simule la propagation du Media Virus à travers le réseau."""
    G = reseau.copy()
    virus = RushkoffVirus(**virus_zero)
    
    # Patient Zéro (Un nœud hautement connecté mais cynique)
    patient_zero = max(G.nodes(), key=lambda x: G.degree(x))
    G.nodes[patient_zero]['etat'] = 'Infecte'
    
    historique = {'Sain': [G.number_of_nodes()-1], 'Infecte': [1]}
    
    for t in range(t_max):
        nouveaux_infectes = []
        virus.muter() # Le virus mute à chaque cycle de replication (cycle de news)
        force_infection = virus.calculer_force_infection()
        
        for node in G.nodes():
            if G.nodes[node]['etat'] == 'Sain':
                # Calcul de la pression virale des voisins
                voisins_infectes = [n for n in G.neighbors(node) if G.nodes[n]['etat'] == 'Infecte']
                if not voisins_infectes:
                    continue
                
                pression_virale = len(voisins_infectes) * force_infection
                
                # LE BYPASS DU SYSTEME IMMUNITAIRE (Le Glitch Reznor)
                # Si le virus est trop "bruyant" (industriel), il choque et passe sous le radar
                # de l'analyse rationnelle (l'immunité).
                seuil_immunite = G.nodes[node]['immunite']
                
                if pression_virale > seuil_immunite:
                    nouveaux_infectes.append(node)
                    
        for n in nouveaux_infectes:
            G.nodes[n]['etat'] = 'Infecte'
            
        # Stats
        sains = sum(1 for n in G.nodes() if G.nodes[n]['etat'] == 'Sain')
        infectes = sum(1 for n in G.nodes() if G.nodes[n]['etat'] == 'Infecte')
        historique['Sain'].append(sains)
        historique['Infecte'].append(infectes)
        
    return historique, G
```

```python
def lancer_simulation(resonance=0.5, mutation=0.3, bruit_reznor=0.8):
    G = creer_systeme_immunitaire(800, 3)
    payload = {
        'resonance_narrative': resonance,
        'taux_mutation': mutation,
        'bruit_industriel': bruit_reznor
    }
    
    hist, G_final = propager_virus(G, payload, t_max=30)
    
    # Visualisation
    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(hist['Infecte'], color='#ff3333', lw=2, label='Hôtes Infectés (Media Virus)')
    ax.fill_between(range(len(hist['Infecte'])), hist['Infecte'], color='#ff3333', alpha=0.2)
    ax.set_title(f"RUSHKOFF GENERATOR // Infection Narrative | Bruit: {bruit_reznor:.1f}", 
                 color='#00ff00', fontsize=14, fontweight='bold')
    ax.set_xlabel("Cycles de Réplication (Temps Médiatique)")
    ax.set_ylabel("Nombre d'Hôtes")
    ax.legend()
    plt.show()

# Widget interactif pour Jupyter
interact(lancer_simulation, 
         resonance=FloatSlider(min=0, max=1, step=0.1, value=0.4, description='Narration'),
         mutation=FloatSlider(min=0, max=1, step=0.1, value=0.5, description='Mutation'),
         bruit_reznor=FloatSlider(min=0, max=2, step=0.1, value=1.2, description='Glitch (Reznor)'));
```
